```python
df_cancer = pd.DataFrame(data=np.c_[cancer.data, cancer.target], columns=np.append(cancer.feature_names, ['target']))
df_cancer.head()
```

# Visualize Data

```python
sns.pairplot(df_cancer, hue='target', vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness'])
```

```python
sns.countplot(df_cancer['target'])
```

```python
sns.scatterplot(data=df_cancer, x='mean area', y='mean smoothness', hue='target')
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(df_cancer.corr(), annot=True)
```

# Model Training

```python
x = df_cancer.iloc[:, 0:30]
y = df_cancer.iloc[:, 30]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(x_train, y_train)
```

# Model evaluation using test datasets

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

# Improving the Model

```python
#Before scaling
sns.scatterplot(data=x_train, x='mean area', y='mean smoothness', hue=y_train)
```

```python
#After scaling
x_train_scaled = (x_train-x_train.min()) / (x_train-x_train.min()).max()
sns.scatterplot(data=x_train_scaled, x='mean area', y='mean smoothness', hue=y_train)
```

```python
x_test_scaled = (x_test-x_test.min()) / (x_test.max()-x_test.min())
```

```python
classifier.fit(x_train_scaled, y_train)
```

```python
y_pred = classifier.predict(x_test_scaled)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```
